```python
# Step 2: Import EPI
from epi_recorder import record
import time
import random
```

### Step 3: Record a Workflow

We will wrap a simple mock AI task with the `record` context manager.

```python
print("ðŸš€ Starting Secured Workflow...")

# Create an evidence package named 'investor_demo.epi'
with record("investor_demo.epi", 
            workflow_name="Investor Validation Demo",
            goal="Showcase EPI verification capabilities",
            tags=["demo", "production"]) as session:
    
    # Simulate some work
    steps = ["Loading Model", "Processing Input", "Generating Output"]
    for step in steps:
        print(f"... {step}")
        time.sleep(0.5)
        session.log_step(step.lower().replace(" ", "_"), {"status": "complete"})
    
    print("âœ… Workflow Finished!")
```

```python
# Step 4: Verify the Evidence Package
!epi verify investor_demo.epi
```
